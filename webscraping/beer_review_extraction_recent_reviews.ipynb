```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
import langdetect
from time import sleep
from random import randint
import json
```

```python
url = f"https://www.ratebeer.com/beer-ratings/0/{1}/"
page = requests.get(url)

soup = bs(page.content, "html.parser")
```

```python
list_reviews = []

for index in range(1,660):
    try:
        sleep(randint(1,5))
        print(f"extracting page {index}")
        url = f"https://www.ratebeer.com/beer-ratings/0/{index}/"
        page = requests.get(url)
        
        soup = bs(page.content, "html.parser")

        beer_table = soup.find_all("table", {"class": "table"})
        beer_reviews = beer_table[0].find_all("div", {"style": "color:#666;"})
        for i in beer_reviews:
            review = i.text
            if langdetect.detect(review) == "en":
                list_reviews.append(review)
    except Exception as e:
        pass
```

```python
dict_reviews = {
    "reviews": list_reviews
}
```

```python
with open("beer_reviews.json", "w") as outfile:
    json.dump(dict_reviews, outfile)
```
